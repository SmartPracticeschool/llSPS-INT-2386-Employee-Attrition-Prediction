# X  contains features and y contains target variable

```python
df1=df.drop(labels="Attrition",axis=1)
df1.head()
```

```python
x=df1
y=df.iloc[:,1]
```

```python
df1.columns
```

# Splitting into training and test

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

# Random Forest Classifier

# Feature Selection using Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import SelectFromModel
```

```python
feat_labels=['Age', 'BusinessTravel', 'DailyRate', 'Department', 'DistanceFromHome',
       'Education', 'EducationField', 'EmployeeCount', 'EmployeeNumber',
       'EnvironmentSatisfaction', 'Gender', 'HourlyRate', 'JobInvolvement',
       'JobLevel', 'JobRole', 'JobSatisfaction', 'MaritalStatus',
       'MonthlyIncome', 'MonthlyRate', 'NumCompaniesWorked', 'Over18',
       'OverTime', 'PercentSalaryHike', 'PerformanceRating',
       'RelationshipSatisfaction', 'StandardHours', 'StockOptionLevel',
       'TotalWorkingYears', 'TrainingTimesLastYear', 'WorkLifeBalance',
       'YearsAtCompany', 'YearsInCurrentRole', 'YearsSinceLastPromotion',
       'YearsWithCurrManager']
```

```python
# Create a random forest classifier
clf = RandomForestClassifier(n_estimators=10000, random_state=1, n_jobs=-1)
```

```python
# Train the classifier
clf.fit(x_train, y_train)
```

```python
# Print the name and gini importance of each feature
k=[]
for feature in zip(feat_labels, clf.feature_importances_):
    k.append(feature)
    print(feature)
```

```python
a=[]
b=[]
for i in k:
    a.append(i[0])
    b.append(i[1])
   
```

```python
plt.bar(a,b)
plt.show()
```

```python
# features that have an importance of more than 0.05
sfm = SelectFromModel(clf, threshold=0.05)
```

```python
sfm.fit(x_train, y_train)
```

```python
# Printing the names of the most important features
for feature_list_index in sfm.get_support(indices=True):
    print(feat_labels[feature_list_index])
```

```python
# Transforming the data to create a new dataset containing only the most important features
X_important_train = sfm.transform(x_train)
X_important_test = sfm.transform(x_test)
```

```python
X_important_train
```

# Training

```python
# Creating a new random forest classifier for the most important features
clf_important = RandomForestClassifier(n_estimators=10000, random_state=1, n_jobs=-1)

# Training the new classifier
clf_important.fit(X_important_train, y_train)
```

# Prediction

```python
# Predicting for test data
y_important_pred = clf_important.predict(X_important_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

# Evaluation

```python
# Viewing The Accuracy 
accuracy_score(y_test, y_important_pred)
```

```python
print(classification_report(y_test, y_important_pred))
```

```python
print(confusion_matrix(y_test, y_important_pred))
```
